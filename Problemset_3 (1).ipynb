## Question 1

```python
import pandas as pd

# Step 2: Load the dataset
# Step 3: Store it in a variable named 'users'
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep='|')

# Step 4: Calculate average age for each occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean()

# Step 5: Determine the ratio of males per occupation and sort the values
def male_ratio(group):
    return (group['gender'] == 'M').sum() / group['gender'].count()

male_ratio_per_occupation = users.groupby('occupation').apply(male_ratio)
male_ratio_per_occupation = male_ratio_per_occupation.sort_values(ascending=False)

# Step 6: Find the minimum and maximum ages for each occupation
age_range_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])

# Step 7: Calculate average age for each combination of occupation and gender
mean_age_per_combination = users.groupby(['occupation', 'gender'])['age'].mean()

# Step 8: Calculate the percentage of women and men for each occupation
def gender_percentage(group):
    total = group['gender'].count()
    return {'percentage_women': (group['gender'] == 'F').sum() / total * 100,
            'percentage_men': (group['gender'] == 'M').sum() / total * 100}

gender_percentage_per_occupation = users.groupby('occupation').apply(gender_percentage)

# Print or display the results as needed
print("Mean Age per Occupation:\n", mean_age_per_occupation)
print("\nMale Ratio per Occupation:\n", male_ratio_per_occupation)
print("\nAge Range per Occupation:\n", age_range_per_occupation)
print("\nMean Age per Occupation and Gender:\n", mean_age_per_combination)
print("\nGender Percentage per Occupation:\n", gender_percentage_per_occupation)
```

## Question 2

```python
import pandas as pd

# Step 2: Import the dataset
# step 3: Assign it to a variable called euro12
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"
euro12 = pd.read_csv(url)

# Step 4: Select only the Goal column
goals = euro12['Goals']

# Step 5: Count the number of teams participating in Euro2012
num_teams = euro12['Team'].nunique()

# Step 6: Get the number of columns in the dataset
num_columns = euro12.shape[1]

# Step 7: Create a DataFrame with selected columns
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]

# Step 8: Sort the teams by Red Cards and then by Yellow Cards
discipline_sorted = discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=False)

# Step 9: Calculate the mean Yellow Cards per Team
mean_yellow_cards_per_team = discipline.groupby('Team')['Yellow Cards'].mean()

# Step 10: Filter teams that scored more than 6 goals
teams_more_than_six_goals = euro12[euro12['Goals'] > 6]

# Step 11: Select teams that start with 'G'
teams_starting_with_G = euro12[euro12['Team'].str.startswith('G')]

# Step 12: Select the first 7 columns
first_7_columns = euro12.iloc[:, :7]

# Step 13: Select all columns except the last 3
columns_except_last_3 = euro12.iloc[:, :-3]

# Step 14: Present Shooting Accuracy for England, Italy, and Russia
shooting_accuracy_selected_teams = euro12.loc[euro12['Team'].isin(['England', 'Italy', 'Russia']), ['Team', 'Shooting Accuracy']]

# Print or display the results as needed
print("Number of Teams Participated in Euro2012:", num_teams)
print("Number of Columns in the Dataset:", num_columns)
print("\nSelected Columns (Team, Yellow Cards, Red Cards):\n", discipline)
print("\nTeams Sorted by Red Cards and then Yellow Cards:\n", discipline_sorted)
print("\nMean Yellow Cards per Team:\n", mean_yellow_cards_per_team)
print("\nTeams that Scored More Than 6 Goals:\n", teams_more_than_six_goals)
print("\nTeams Starting with 'G':\n", teams_starting_with_G)
print("\nFirst 7 Columns:\n", first_7_columns)
print("\nColumns Except the Last 3:\n", columns_except_last_3)
print("\nShooting Accuracy for England, Italy, and Russia:\n", shooting_accuracy_selected_teams)
```

```python
import pandas as pd
import numpy as np

# Step 2: Create 3 different Series
series_1 = pd.Series(np.random.randint(1, 5, size=100))  # Random number from 1 to 4
series_2 = pd.Series(np.random.randint(1, 4, size=100))  # Random number from 1 to 3
series_3 = pd.Series(np.random.randint(10000, 30001, size=100))  # Random number from 10,000 to 30,000

# Step 3: Create a DataFrame by joining the Series by column
data = pd.concat([series_1, series_2, series_3], axis=1)
data.columns = ['bedrs', 'bathrs', 'price_sqr_meter']

# Step 5: Create a one column DataFrame with the values of the 3 Series
bigcolumn = pd.concat([series_1, series_2, series_3], ignore_index=True)

# Step 6: Check if the index goes only until 99
is_index_until_99 = bigcolumn.index.equals(pd.RangeIndex(start=0, stop=100))

# Step 7: Reindex the DataFrame
bigcolumn = bigcolumn.reset_index(drop=True)

# Print or display the results as needed
print("DataFrame created by joining the Series:\n", data)
print("\nOne column DataFrame 'bigcolumn':\n", bigcolumn)
print("\nIs 'bigcolumn' index only until 99?", is_index_until_99)
```

## Question 4

```python
import pandas as pd

# Step 2: Import the dataset
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data"
data = pd.read_csv(url, sep='\s+', parse_dates=[[0, 1, 2]])

# Step 4: Fix the year 2061
def fix_year(year):
    if year.year > 2000:
        return year - pd.DateOffset(years=100)
    return year

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_year)

# Step 5: Set the right dates as the index
data.set_index('Yr_Mo_Dy', inplace=True)

# Step 6: Count missing values for each location
missing_values_per_location = data.isnull().sum()

# Step 7: Count non-missing values in total
total_non_missing_values = data.notnull().sum().sum()

# Step 8: Calculate the mean windspeeds over all the locations and times
mean_windspeed = data.mean().mean()

# Step 9: Calculate statistics for each location
loc_stats = data.describe()

# Step 10: Calculate statistics for each day
day_stats = data.resample('D').agg(['min', 'max', 'mean', 'std'])

# Step 11: Average windspeed in January for each location
january_avg_per_location = data.resample('Y').mean()
january_avg_per_location = january_avg_per_location[january_avg_per_location.index.month == 1]

# Print or display the result for Step 11
print("\nAverage Windspeed in January for Each Location:\n", january_avg_per_location)


# Step 12: Downsample to yearly frequency
yearly_data = data.resample('Y').mean()

# Step 13: Downsample to monthly frequency
monthly_data = data.resample('M').mean()

# Step 14: Downsample to weekly frequency
weekly_data = data.resample('W-TUE').mean()  # Assume the first week starts on January 2, 1961

# Step 15: Calculate statistics for each week (first 52 weeks)
weekly_stats = weekly_data.head(52).agg(['min', 'max', 'mean', 'std'])

# Print or display the results as needed
print("Missing Values per Location:\n", missing_values_per_location)
print("\nTotal Non-Missing Values:", total_non_missing_values)
print("\nMean Windspeeds over All Locations and Times:", mean_windspeed)
print("\nStatistics for Each Location:\n", loc_stats)
print("\nStatistics for Each Day:\n", day_stats)
print("\nAverage Windspeed in January for Each Location:\n", january_avg_per_location)
print("\nYearly Downsampled Data:\n", yearly_data)
print("\nMonthly Downsampled Data:\n", monthly_data)
print("\nWeekly Downsampled Data:\n", weekly_data)
print("\nStatistics for Each Week (First 52 Weeks):\n", weekly_stats)
```

```python
import pandas as pd

# Step 2: Import the dataset
# step 3: Assign it to a variable called chipo
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t')

# Step 4: See the first 10 entries
first_10_entries = chipo.head(10)

# Step 5: Number of observations
num_observations = chipo.shape[0]

# Step 6: Number of columns
num_columns = chipo.shape[1]

# Step 7: Name of all columns
columns_names = chipo.columns.tolist()

# Step 8: Indexing type
indexing_type = chipo.index

# Step 9: Most-ordered item
most_ordered_item = chipo['item_name'].value_counts().idxmax()

# Step 10: Number of most-ordered item
num_most_ordered_item = chipo[chipo['item_name'] == most_ordered_item]['quantity'].sum()

# Step 11: Most ordered item in choice_description
most_ordered_choice = chipo['choice_description'].value_counts().idxmax()

# Step 12: Total number of items ordered
total_items_ordered = chipo['quantity'].sum()

# Step 13: Convert item price to float
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))

# Step 14: Revenue for the period
revenue = (chipo['quantity'] * chipo['item_price']).sum()

# Step 15: Number of orders made
num_orders = chipo['order_id'].nunique()

# Step 16: Average revenue per order
avg_revenue_per_order = revenue / num_orders

# Step 17: Number of different items sold
num_different_items = chipo['item_name'].nunique()

# Print or display the results as needed
print("First 10 Entries:\n", first_10_entries)
print("\nNumber of Observations:", num_observations)
print("\nNumber of Columns:", num_columns)
print("\nColumns Names:", columns_names)
print("\nIndexing Type:", indexing_type)
print("\nMost-Ordered Item:", most_ordered_item)
print("\nNumber of Most-Ordered Item:", num_most_ordered_item)
print("\nMost-Ordered Choice Description:", most_ordered_choice)
print("\nTotal Items Ordered:", total_items_ordered)
print("\nRevenue for the Period:", revenue)
print("\nNumber of Orders Made:", num_orders)
print("\nAverage Revenue Amount per Order:", avg_revenue_per_order)
print("\nNumber of Different Items Sold:", num_different_items)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import the dataset
file_path = "C:/Users/Dell/Downloads/roman-emperor-reigns.csv"
data = pd.read_csv(file_path)

# Step 2: Calculate the number of emperors assassinated and not assassinated
assassinated_count = data[data['Cause_of_Death'] == 'Assassinated'].shape[0]
not_assassinated_count = data[data['Cause_of_Death'] != 'Assassinated'].shape[0]

# Step 3: Create a pie chart
plt.figure(figsize=(8, 8))

# Plot the data
plt.pie([assassinated_count, not_assassinated_count], labels=['Assassinated', 'Not Assassinated'],
        autopct='%1.1f%%', startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Add title
plt.title('Fraction of Roman Emperors Assassinated')

# Show the plot
plt.show()
```
